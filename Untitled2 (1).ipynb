Q 17.A Explain the difference between apply() and map() in Pandas
- 1. map()
Works mainly on Series (1D data).

Used to map values of a Series according to an input correspondence (like a dictionary or a function).

Can be used for element-wise transformations or value substitutions.

Does not work on DataFrames directly.

2. apply()
Works on both Series and DataFrames.

Applies a function along an axis of the DataFrame or to each element of a Series.

More flexible: can apply row-wise or column-wise functions for DataFrames.

Suitable for complex row/column transformations.

Q 21.Why is NumPy‚Äôs array slicing faster than Python‚Äôs list slicing
- Contiguous Memory Storage

NumPy arrays store data in a continuous block of memory (like C arrays).

Python lists store references (pointers) to objects scattered in memory.

This contiguous layout allows faster access and slicing in NumPy, as it can directly compute memory offsets.

- Homogeneous Data Types

NumPy arrays hold elements of the same data type (e.g., all floats).

Python lists can contain heterogeneous types, requiring extra overhead for type checking during slicing.

- View vs Copy

NumPy slicing usually returns a view (a lightweight reference) of the original array without copying data.

Python list slicing creates a new list and copies references to the sliced elements, which is more time-consuming.

- Optimized Low-Level Implementation

NumPy operations are implemented in compiled C code optimized for performance.

Python list slicing is implemented in Python‚Äôs interpreter, which is slower.

4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix Heatmap")
plt.show()
```
